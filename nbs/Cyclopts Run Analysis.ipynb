```python
times = defaultdict(list)
for x in results.iterrows():
    times[x['solver']].append((x['instid'], x['time']))

objectives = defaultdict(list)
for x in results.iterrows():
    objectives[x['solver']].append((x['instid'], x['objective']))
```

```python
def plot_xy(xname, xhandle, yvals=times, ylabel='Solution Time (s)',
            title_type='Solution Time'):
    vals = {x['instid']: x[xhandle] for x in props.iterrows()}
    
    markers = iter(['x', 's', 'o'])
    colors = iter(['g', 'b', 'r'])

    kinds = times.keys()
    plts = {}
    for s, l in yvals.iteritems():
        x = [vals[i[0]] for i in l]
        y = [i[1] for i in l]
        plts[s] = plt.scatter(x, y, c=colors.next(), marker=markers.next(), label=s)    
    plt.xlabel('Number of {0}s'.format(xname))
    plt.ylabel(ylabel)
    plt.title(r'Distribution of {1} as a Function of {0} Density'.format(xname, title_type))
    plt.xlim(0)
    plt.ylim(0)
    plt.legend([p for _, p in plts.items()], [s for s, _ in plts.items()], loc=2)
```

```python
xname = 'Request Node'
xhandle = 'n_u_nodes'
plot_xy(xname, xhandle)
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_xyz(xname, xhandle, yname, yhandle, xtbl=props, 
             ytbl=props, zvals=times, zlabel='Solution Time (s)',
             title=r'Distribution of Solution Time'):
    xvals = {x['instid']: x[xhandle] for x in xtbl.iterrows()}
    yvals = {x['instid']: x[yhandle] for x in ytbl.iterrows()}
   
    markers = iter(['x', 's', 'o'])
    colors = iter(['g', 'b', 'r'])

    fig = plt.figure()
    ax = plt.axes(projection='3d')
    kinds = times.keys()
    plts = {}
    for s, l in zvals.iteritems():
        x = [xvals[i[0]] for i in l]
        y = [yvals[i[0]] for i in l]
        z = [i[1] for i in l]
        plts[s] = ax.scatter(x, y, z, c=colors.next(), marker=markers.next(), label=s)    

    ax.set_xlabel('Number of {0}s'.format(xname))
    ax.set_ylabel('Number of {0}s'.format(yname))
    ax.set_zlabel(zlabel)
    ax.set_xlim(0)
    ax.set_ylim(0)
    ax.set_zlim(0)
    plt.title(title)
#plt.legend([p for _, p in plts.items()], [s for s, _ in plts.items()], loc=2)
```

```python
rel_diff = lambda x, y: (x - y) / y
diff = lambda x, y: (x - y)

def solver_diff(vals, base, baseline, f=rel_diff):
    diffs = {}
    for s, l in vals.iteritems():
        if s == base:
            continue
        diffs[s] = [f(x[1], baseline[x[0]]) for x in l]
    return diffs
```

```python
vals = objectives
base = 'cbc'
baseline = {x[0]: x[1] for x in vals[base]}
d = solver_diff(vals, base, baseline, f=rel_diff)

n, bins, patches = plt.hist(d.values(), 40, label=d.keys())
plt.legend()
plt.xlabel('Relative Difference'.format(name))
plt.ylabel('Number of Instances')
plt.title(r'Distribution of Objective Difference')
    
```

```python
vals = times
base = 'cbc'
baseline = {x[0]: x[1] for x in vals[base]}
d= solver_diff(vals, base, baseline, f=rel_diff)

n, bins, patches = plt.hist(d.values(), 20, label=d.keys())
plt.legend()
plt.xlabel('Relative Difference'.format(name))
plt.ylabel('Number of Instances')
plt.title(r'Distribution of Time Difference')
```

```python
grps = f.root.Instances.ExchangeGroups
samplers = f.root.ReactorRequestSampler
flows = f.root.Instances.ExchangeInstSolutions
props = f.root.Instances.ExchangeInstProperties

req_grp_qtys = defaultdict(float)
total_mass_flow = defaultdict(float)
total_node_qty = {}
inst_to_param = {}
soln_to_inst = {}

for x in props.iterrows():
    inst_to_param[x['instid']] = x['paramid']

for x in grps.iterrows():
    if x['kind'] == 1: # supply == 0, req == 1
        req_grp_qtys[x['instid']] += x['caps'][0]
    
for x in samplers.iterrows():
    total_node_qty[x['paramid']] = x['assem_per_req_avg'] * x['n_request_avg']

# test modeling assumption that the group mass constraint is equal to the expected total request quantity
obs = [qty - total_node_qty[inst_to_param[iid]] for iid, qty in req_grp_qtys.items()]
exp = [0] * len(total_node_qty)
np.testing.assert_array_equal(exp, obs)
    
for x in flows.iterrows():
    total_mass_flow[x['solnid']] += x['flow']
    soln_to_inst[x['solnid']] = x['instid']

# test truth that total flows are at most equal to the group mass constraint 
tol = 1e-8
obs = np.array([total_mass_flow[sid] - req_grp_qtys[iid] for sid, iid in soln_to_inst.items()])
pos_idx = np.where(obs > tol)
np.testing.assert_equal(0, len(pos_idx[0]))
```
