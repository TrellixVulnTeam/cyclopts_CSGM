Result Characterization
-----------------------

```python
times = defaultdict(list)
for x in results.iterrows():
    times[x['solver']].append((x['instid'], x['time']))

objectives = defaultdict(list)
for x in results.iterrows():
    objectives[x['solver']].append((x['instid'], x['objective']))
```

```python
def plot_xy(xname, xhandle, yvals=times):
    vals = {x['instid']: x[xhandle] for x in props.iterrows()}
    
    markers = iter(['x', 's', 'o'])
    colors = iter(['g', 'b', 'r'])

    kinds = times.keys()
    plts = {}
    for s, l in yvals.iteritems():
        x = [vals[i[0]] for i in l]
        y = [i[1] for i in l]
        plts[s] = plt.scatter(x, y, c=colors.next(), marker=markers.next(), label=s)    
    plt.xlabel('Number of {0}s'.format(xname))
    plt.ylabel('Solution Time (s)')
    plt.title(r'Distribution of Solution Time as a Function of {0} Density'.format(xname))
    plt.xlim(0)
    plt.ylim(0)
    plt.legend([p for _, p in plts.items()], [s for s, _ in plts.items()], loc=2)
```

```python
xname = 'Request Node'
xhandle = 'n_u_nodes'
plot_xy(xname, xhandle)
```

```python
xname = 'Arc'
xhandle = 'n_arcs'
plot_xy(xname, xhandle)
```

```python
xname = 'Constraint'
xhandle = 'n_constrs'
plot_xy(xname, xhandle)
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_xyz(xname, xhandle, yname, yhandle, xtbl=props, ytbl=props, zvals=times):
    xvals = {x['instid']: x[xhandle] for x in xtbl.iterrows()}
    yvals = {x['instid']: x[yhandle] for x in ytbl.iterrows()}
   
    markers = iter(['x', 's', 'o'])
    colors = iter(['g', 'b', 'r'])

    fig = plt.figure()
    ax = plt.axes(projection='3d')
    kinds = times.keys()
    plts = {}
    for s, l in zvals.iteritems():
        x = [xvals[i[0]] for i in l]
        y = [yvals[i[0]] for i in l]
        z = [i[1] for i in l]
        plts[s] = ax.scatter(x, y, z, c=colors.next(), marker=markers.next(), label=s)    

    ax.set_xlabel('Number of {0}s'.format(xname))
    ax.set_ylabel('Number of {0}s'.format(yname))
    ax.set_zlabel('Solution Time (s)')
    ax.set_xlim(0)
    ax.set_ylim(0)
    ax.set_zlim(0)
    plt.title(r'Distribution of Solution Time')
#plt.legend([p for _, p in plts.items()], [s for s, _ in plts.items()], loc=2)
```

```python
xname = 'Arc'
xhandle = 'n_arcs'
yname = 'Constraint'
yhandle = 'n_constrs'
plot_xyz(xname, xhandle, yname, yhandle)
```

```python
xname = 'Arc'
xhandle = 'n_arcs'
yname = 'Exclusivity'
yhandle = 'excl_frac'
plot_xyz(xname, xhandle, yname, yhandle)
```

```python
rel_diff = lambda x, y: (x - y) / y

def diff_hist(vals, base, baseline):
    kinds = []
    diffs = []
    for s, l in vals.iteritems():
        if s == base:
            continue
        diffs.append([rel_diff(x[1], baseline[x[0]]) for x in l])
        kinds.append(s)
    n, bins, patches = plt.hist(diffs, 30, label=kinds)
    plt.legend()
#plt.xlabel('Number of {0}s'.format(name))
#plt.ylabel('Number of Instances')
#plt.title(r'Distribution of {0} Quantities in Problem Instances'.format(name))
    
```

```python
vals = objectives
base = 'cbc'
baseline = {x[0]: x[1] for x in vals[base]}
diff_hist(vals, base, baseline)
```
